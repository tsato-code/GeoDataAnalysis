```python
import matplotlib.pyplot as plt
import shapefile
import warnings
warnings.filterwarnings('ignore')
```

```python
readfile = "data/taxi_zones/taxi_zones.shp"
sf = shapefile.Reader(readfile)
```

```python
fig = plt.figure()
plt.axis('off')
ax = fig.add_subplot(111)
for shape in sf.shapes():
    points = shape.points
    ap = plt.Polygon(points, fill=True, edgecolor="k")
    ax.add_patch(ap)
ax.autoscale_view()
plt.show()
```
